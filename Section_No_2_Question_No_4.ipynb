```python
# Defining the indices for each fold

split_fold_1 = (0, 1000)
split_fold_2 = (1000, 2000)
split_fold_3 = (2000, 3000)
split_fold_4 = (3000, 4000)
split_fold_5 = (4000, 5000)
```

```python
# Different values of k

values_of_k = [1, 3, 5, 7, 10]
```

```python
# Performing k cross fold validation where k = 5

for value in values_of_k:
  for split_fold in range(1, 6):
    start_of_test, end_of_test = globals()[f"split_fold_{split_fold}"] # Reference for globals() (https://www.programiz.com/python-programming/methods/built-in/globals)
    test_set = data_email[start_of_test:end_of_test]
    train_set = np.concatenate([data_email[:start_of_test], data_email[end_of_test:]])


  for point in test_set:

    # Calculating Euclidean distance between test instance and each of training instances
    euclidean_distances = np.vectorize(np.linalg.norm(np.vectorize(int(train_set[:, 1:])) - np.vectorize((point[1:])), axis=1))

    # Finding the index of Nearest Neighbor where k being different values in the list initialized above.
    index_of_nearest_neighbor = np.argmin(euclidean_distances)

    # Prediction and the true label extraction
    predicted_label = train_set.loc[index_of_nearest_neighbor]
    true_label = point

    # Evaluating the accuracy, precision and recall scores for each fold.

    accuracy = accuracy_score(true_label, predicted_label)

    print("Fold.No:- ", split_fold)
    print("\n")
    print("Accuracy score:- ", accuracy)
    print("\n")
```
